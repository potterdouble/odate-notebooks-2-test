### Binder Specific Memos

Aside from timing out after 10 minutes of inactivity, there's a couple other Binder specific things to keep in mind.

#### Memory Management

At the bottom of the JupyterLab interface, you'll see something like this:

![Memory indicator in Binder](./data/mem-reg.png)

Binder allows you to program in your browser by creating a virtual computer that hosts JupyterLab alongside the files provided in your GitHub repository. Like a regular computer, a Binder environment only has so much virtual memory that can be used for computation. If you run out of memory, your Binder instance will crash meaning you'll have to reopen the Binder link in a new tab and begin again. If you're close to running out of memory, you'll be warned when this section turns red:

![Running out of virtual memory in Binder](./data/mem-red.png)

To prevent this from happening, you can make a habit of removing the session kernels that are created each time you start up a new notebook once you're finished with the exercise. To do this, close the notebook you're done with then click the small "CPU" icon that also appears at the bottom left of the interface:

![Open kernel manager](./data/open-kernel-manager.png)

This will open a new tab where the file browser usually is, to the left of the screen. Here, when you hover over the kernel (which shares the name of the notebook you were just working with) there will be the option to click the "X" and shut it down.

![Shut down the kernel](./data/shutdown-kernel.png)

After shutting down the kernel, you'll see the memory free up which means you can now continue on to the next exercise!
