```python
# important dependencies
import spacy
nlp = spacy.load('en_core_web_lg')

import nltk
nltk.download('punkt')

from convokit import Corpus, download
corpus = Corpus(filename=download("movie-corpus"))

# This bit creates a lookup dictionary for items with the ID movie_lines
movie_lines = {}
for line in open(r"C:\Users\Kegan\.convokit\saved-corpora\movie-corpus\utterances.jsonl",
                 encoding="latin1"):
    line = line.strip()
    parts = line.split(" +++$+++ ")
    if len(parts) == 5:
        movie_lines[parts[0]] = parts[4]
    else:
        movie_lines[parts[0]] = ""
```

```python
import json
responses = {}
for line in open(r"C:\Users\Kegan\.convokit\saved-corpora\movie-corpus\conversations.json",
                 encoding="latin1"):
    line = line.strip()
    parts = line.split(" +++$+++ ")
    line_ids = json.loads(parts[0].replace("'", '"'))
    for first, second in zip(line_ids[:-1], line_ids[1:]):
        responses[first] = second
```

```python
import random
for pair in random.sample(responses.items(), 5):
    print("A:", movie_lines[pair[0]])
    print("B:", movie_lines[pair[1]])
    print()
```
